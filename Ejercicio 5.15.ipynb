Realizar un programa que calcule la derivada de la función:
\begin{equation}
f(x)=1+\frac{1}{2}tanh(2x)
\end{equation}

Utilizando el método de derivadas hacia adelante y derivadas centrales, Por otra parte, compare dicha derivada numérica con la análica mediante un gráfico, además grafique el porcentaje de error entre las mísmas.



```python
import math as mt
import matplotlib.pyplot as plt
def f(x):
    return 1+(1/2)*mt.tanh(2*x)   # Función que se desea derivar 

def g(x):
    return 1-mt.tanh(2*x)**2    # Derivada de la función 

a=-2
b=2
N=100
h=(b-a)/N

fp=[]
for i in range(N):
    fp.append(f(a+i*h))

plt.plot(fp,'o') 
```

Derivadas numéricas por el método: $\textbf{Derivadas}$ $\textbf{hacia}$ $\textbf{adelante}$

```python
dfn=[]
dfa=[]

for i in range(N-1):
    dfn.append((fp[i+1]-fp[i])/h)
    dfa.append(g(a+i*h))
```

Comparamos la derivada Analítica y Númerica

```python
plt.plot(dfa)
plt.plot(dfn,'o')
```

```python
diff=[]

for i in range(N-1):
    tmp=1-(dfn[i]/dfa[i])          # Diferencia entre la derivada numérica y analítica (Porcentaje de Error)
    diff.append(tmp*100)
```

```python
plt.plot(diff,'o')
```

Derivadas Númericas por el método de $\textbf{Derivadas}$ $\textbf{Centrales}$

```python
dfn=[]
dfa=[]

for i in range(1,N-1):
    dfn.append((fp[i+1]-fp[i-1])/(2*h)) 
    dfa.append(g(a+i*h))
    
```

Comparación entre la derivada analítica y númerica

```python
plt.plot(dfa)
plt.plot(dfn,'o')
```

```python
diff=[]

for i in range(N-2):                    # Diferencia entre la derivada numérica y analítica (Porcentaje de Error)
    tmp=1-(dfn[i]/dfa[i])
    diff.append(tmp*100)
```

```python
plt.plot(diff,'o')
```

Se concluye que el método de derivadas centrales es más preciso que el método de derivas hacia adelante. Ya que la diferencia porcentual de error es menor en el método de diferenciación central.
